```python
import numpy as np
import tensorflow as tf
import json

# Load TFLite model and allocate tensors.
interpreter = tf.lite.Interpreter(model_path="squeezenet.tflite")
interpreter.allocate_tensors()
```

```python
def checkKey(dict, key):
    
    if key in dict.keys():
        return True
    else:
        return False
```

```python
# Get input and output tensors.
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
```

```python
profiler = interpreter.get_tensor_details()
```

```python
for tensor in profiler:
    print("----- "+tensor["name"]+" -----")
    print("index: " + str(tensor["index"]))
    print("shape: " + str(tensor["shape"]))
```

```python
layers = {}

for tensor in profiler:
    path = tensor["name"].split("/")
    if (len(path) == 1):
        layers[path[0]] = tensor["shape"].tolist()
    elif(len(path) == 2):
        if(not checkKey(layers, path[0])):
            layers[path[0]] = {}
        layers[path[0]][path[1]] = tensor["shape"].tolist()
    else:
        if(not checkKey(layers, path[0])):
            layers[path[0]] = {}
        if(not checkKey(layers[path[0]], path[1])):
            layers[path[0]][path[1]] = {}
        layers[path[0]][path[1]][path[2]] = tensor["shape"].tolist()
```

```python
with open('squezzenet_layers.txt', 'w') as outfile:
    json.dump(layers, outfile)
```
